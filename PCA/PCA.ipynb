```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.decomposition import PCA
```

```python
import numpy as np
```

```python

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Implement PCA
pca = PCA()
pca.fit(scaled_data)

# Determine optimal number of components
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# Choose number of components (e.g., based on the plot above)
n_components = 3  # Choose number of components

# Transform original dataset into principal components
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(n_components)])
```

```python
from sklearn.cluster import KMeans
```

```python
import sklearn.metrics as metrics
```

```python
# Apply K-means clustering to original dataset
kmeans = KMeans(n_clusters=3, random_state=42)
clusters_original = kmeans.fit_predict(scaled_data)

# Visualize clustering results
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'], hue=clusters_original, palette='viridis', legend='full')
plt.title('Clustering with Original Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Evaluate clustering performance (example using silhouette score)
silhouette_score = metrics.silhouette_score(scaled_data, clusters_original, metric='euclidean')
print(f"Silhouette Score for original data: {silhouette_score:.2f}")
```

```python
# Apply K-means clustering to PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)
clusters_pca = kmeans_pca.fit_predict(principal_components)

# Visualize clustering results with PCA data
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'], hue=clusters_pca, palette='viridis', legend='full')
plt.title('Clustering with PCA-transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Evaluate clustering performance (example using silhouette score)
silhouette_score_pca = metrics.silhouette_score(principal_components, clusters_pca, metric='euclidean')
print(f"Silhouette Score for PCA-transformed data: {silhouette_score_pca:.2f}")
```

```python
# Compare clustering results
print("\nComparison of clustering results:")
print("Original Data - Cluster Counts:")
print(pd.Series(clusters_original).value_counts())
print("\nPCA-transformed Data - Cluster Counts:")
print(pd.Series(clusters_pca).value_counts())

# Discuss similarities and differences in clustering results
# Reflect on the impact of dimensionality reduction on clustering performance
# Analyze trade-offs between using PCA and clustering directly on the original dataset
```
